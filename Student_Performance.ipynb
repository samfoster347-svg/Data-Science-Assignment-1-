```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
StudentPerformance = pd.read_csv('StudentsPerformance.csv')

#print(StudentPerformance.isnull().sum())
#print(StudentPerformance.isna().sum()) in neither of these does a missing value appear, so the data seems to be clean already

plt.figure(figsize=(8, 6), dpi=300)

Scores = StudentPerformance[['gender','reading score', 'math score']].copy()
ScoresMelt = Scores.melt(id_vars='gender', var_name='Subject', value_name='Score')

sns.boxplot(y='Score', x='Subject', hue='gender', data=ScoresMelt)

plt.title('V1 - Test scores by Subject and Gender')

plt.show()
```

```python
plt.figure(figsize=(8, 6), dpi=300)
sns.boxplot(y='math score', x='test preparation course', data=StudentPerformance)
plt.title('V2 - Math Score by Test Prep Completion')
plt.show()
```

```python
plt.figure(figsize=(8, 6), dpi=300)
Scores1 = StudentPerformance[['lunch','reading score', 'math score','writing score']].copy()
ScoresMelt1 = Scores1.melt(id_vars='lunch', var_name='score type', value_name='score')
sns.boxplot(y='score', x='score type', hue= 'lunch', data=ScoresMelt1)
plt.title('V3 - Test scores by Subject and Lunch type')
plt.xlabel('Test Score')
plt.ylabel('Score')
plt.legend(title= 'Lunch Type')
plt.show()
```

```python
plt.figure(figsize=(8, 6), dpi=300)
sns.scatterplot(x='math score', y='reading score', hue='writing score', data=StudentPerformance)
plt.title('V4 - Reading Score vs Math Score vs Writing score')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.legend(title= 'Writing Score')
plt.show()
```

```python
plt.figure(figsize=(8, 6), dpi = 300)
sns.lmplot(x='reading score', y='math score', hue = 'test preparation course', data=StudentPerformance)
plt.title('V5- Reading Score vs Math Score by Test Prep Completion')
plt.xlabel('Reading Score')
plt.ylabel('Math Score')
plt.legend(title= 'Test Prep Course Completion')
plt.show()
```
